```python
# Most Runs 2022
deliveries[['batsman', 'batsman_runs']].groupby('batsman').sum().sort_values(by='batsman_runs', ascending=False)[:10]
```

```python
# Most Wickets
deliveries[['bowler', 'isWicketDelivery']].groupby('bowler').sum().sort_values(by='isWicketDelivery', ascending=False).rename(columns={'isWicketDelivery': 'wickets'})[:10]
```

```python
# Most Fours (4s)
deliveries[['batsman', 'batsman_runs']][deliveries['batsman_runs'] == 4].groupby('batsman').count().sort_values(by='batsman_runs', ascending=False).rename(columns={'batsman_runs': 'No. of 4s'})[:10]
```

```python
# Most Fours (4s) in an innings
deliveries[['id', 'batsman', 'batsman_runs']][deliveries['batsman_runs'] == 4].groupby(['id', 'batsman']).count().sort_values(by='batsman_runs', ascending=False).rename(columns={'batsman_runs': 'No. of 4s'})[:10]
```

```python
# Most Sixes (6s)
deliveries[['batsman', 'batsman_runs']][deliveries['batsman_runs'] == 6].groupby('batsman').count().sort_values(by='batsman_runs', ascending=False).rename(columns={'batsman_runs': 'No. of 6s'})[:10]
```

```python
# Most Sixes (6s) In a Innings
deliveries[['id', 'batsman', 'batsman_runs']][deliveries['batsman_runs'] == 6].groupby(['id', 'batsman']).count().sort_values(by='batsman_runs', ascending=False).rename(columns={'batsman_runs': 'No. of 6s'})[:10]
```

```python
# Most Centuries
batsman_runs = deliveries[['id', 'batsman', 'batsman_runs']].groupby(['id', 'batsman']).sum(numeric_only=True)
batsman_runs[batsman_runs['batsman_runs'] >= 100].groupby('batsman').count().sort_values(by='batsman_runs', ascending=False).rename(columns={'batsman_runs': 'No. of 100s'})
```

```python
# Most Fifties
batsman_runs = deliveries[['id', 'batsman', 'batsman_runs']].groupby(['id', 'batsman']).sum(numeric_only=True).sort_values(by='batsman_runs', ascending=False)
fifties = batsman_runs[batsman_runs['batsman_run'] >= 50]
fifties[fifties['batsman_runs'] < 100].groupby('batsman').count().sort_values(by='batsman_runs', ascending=False).rename(columns={'batsman_runs': 'No. of 50s'})[:10]
```

```python
# Q. de Kock's wagonwheel for the innings he scored 140 runs
match = deliveries[deliveries['id'] == 1304112]
deKockInnings = match[match['batsman'] == 'Q de Kock']
wagonwheel = deKockInnings[['batsman_runs', 'total_runs']].groupby('batsman_runs').count().rename(columns={'total_runs': 'runs'})[1:]
wagonwheel.plot.pie(y='runs', autopct='%1.2f%%')
```

```python
# Fastest Fifties
ball_faced = deliveries[['id', 'batsman', 'batsman_runs']].groupby(['id', 'batsman']).count().rename(columns={'batsman_runs': 'balls_faced'})

runs_scored = deliveries[['id', 'batsman', 'batsman_runs']].groupby(['id', 'batsman']).sum()
filterd_runs = runs_scored[runs_scored['batsman_runs'] >= 50]
fifties = filterd_runs[filterd_runs['batsman_runs'] < 100].sort_values(by='batsman_runs', ascending=False)


fifties.merge(ball_faced, on=['id', 'batsman']).sort_values(by='balls_faced', ascending=True)[:10]
```
